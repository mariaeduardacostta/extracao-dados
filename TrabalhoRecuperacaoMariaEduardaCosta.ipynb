```python
import requests
import json
import matplotlib.pyplot as plt
from collections import Counter

url = "https://openlibrary.org/search.json?q=computer+engineering"
response = requests.get(url)
data = json.loads(response.text)

publishers = []
for item in data['docs']:
    if 'publisher' in item:
        publisher = item['publisher']
        for pub in publisher:
          publishers.append(pub)

authors = []
for item in data['docs']:
    if 'author_name' in item:
        author = item['author_name']
        for aut in author:
          authors.append(aut)
          

publisher_count = Counter(publishers)
top_publishers = publisher_count.most_common(10)

author_count = Counter(authors)
top_authors = author_count.most_common(10)

labels, values = zip(*top_publishers)

plt.bar(labels, values)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Editora')
plt.ylabel('Nmero de Livros')
plt.title('Editoras que mais Publicaram Livros de Engenharia')
plt.show()

labels, values = zip(*top_authors)

plt.bar(labels, values)
plt.xticks(rotation=45, ha='right')
plt.xlabel('Autor')
plt.ylabel('Nmero de Livros')
plt.title('Autores que mais publicaram livros de engenharia')
plt.show()
```
